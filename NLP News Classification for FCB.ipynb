#### Tokenize the News text file, we make two news file(news and news1 ) to reduce the file size for the tokenize

```python
import nltk
file_content = open("news",encoding='utf-8').read()
#file_writer = open("TKTENA.txt", 'w' ,encoding='utf-8')
tokens = nltk.word_tokenize(file_content)
print (tokens)
```

### Save tokenized words to a text file (Tokenizetena.txt)

```python
document_text2 = open('news', 'r',encoding='utf-8')
tokenize1 = open('Tokenizetena.txt', 'w',encoding='utf-8')
text_string = document_text2.read()
for word in text_string:
    rf = str(text_string.split())
    tokenize1.write(rf)
    tokenize1.write('\n')
```

#### Count word frequnce to remove highly frequent words

```python
import operator
frequency = {}
#words = open ("tena.txt",encoding='utf-8').read()
#WordFreq = dict()  
for word in tokens:
    count = frequency.get(word,0)
    frequency[word] = count + 1
    
    frequency_list = frequency.keys()
    
    for word in frequency_list:
            WordFreq = sorted(frequency.items(), key= operator.itemgetter(1), reverse=True)

selected=[]
top=15
i=1
for w in WordFreq:
    selected.append(w[0])
    i=i+1
    if i==top+1:
        break

#Print the top kewords selected from the text
print("\nThe selected top "+str(top)+" words are: ")
for s in selected:
    print(s ,frequency[s])
```

#### Open the stemmed news file and extract the news catagories, by counting No. of lines per articles

```python
with open("stemed.txt","r", encoding="utf-8") as f:
    text = f.read()
    news = text.split("\n\n")
    count = {"ስፖርት": 0, "ጤና": 0, "ቢዝነስ": 0, "ቴክኖሎጂ": 0, "ትምህርት": 0, "ፖለቲካ": 0}
    for news_item in news:
        lines = news_item.split("\n")
        print(lines[11])
        file_to_write = open("data/" + lines[11] + '/' + str(count[lines[11]]) + ".txt", "w+", encoding="utf-8")
        count[lines[11]] = count[lines[11]] + 1
        file_to_write.write(news_item)  # python will convert \n to os.linesep
        file_to_write.close()
```

##### Loop through directery list and save to the training list and we use 'glob' lib to loop through all the catagory and directory list we define 

```python
import pandas
import glob
category_list = ["ስፖርት", "ጤና", "ቢዝነስ", "ቴክኖሎጂ", "ትምህርት", "ፖለቲካ", 'utf-8']
directory_list = ["data/ስፖርት/*.txt", "data/ጤና/*.txt", "data/ቢዝነስ/*.txt", "data/ቴክኖሎጂ/*.txt", "data/ትምህርት/*.txt", "data/ፖለቲካ/*.txt", 'utf-8']

text_files = list(map(lambda x: glob.glob(x), directory_list))
text_files = [item for sublist in text_files for item in sublist]

training_data = []

for t in text_files:
    f = open(t, 'r', encoding='utf-8')
    f = f.read()
    t = f.split('\n')
    training_data.append({'data' : t[0] + ' ' + t[1], 'flag' : category_list.index(t[11])})

training_data[0]
```

#### create the pandas data frame and count total articles

```python
training_data = pandas.DataFrame(training_data, columns=['data', 'flag'])
training_data.to_csv("train_data.csv", sep=',', encoding = 'utf-8')
print(training_data.data.shape)
```

###  we use Count Vectorizer which is to count each words in the traing dataset  & save it to 'count_vector.pkl'

```python
import pickle
from sklearn.feature_extraction.text import CountVectorizer

#GET VECTOR COUNT
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(training_data.data)

#SAVE WORD VECTOR
pickle.dump(count_vect.vocabulary_, open("count_vector.pkl","wb"))
```

### Transform Word Vector to TFIDF to count the frequency of words in all documents 

```python
from sklearn.feature_extraction.text import TfidfTransformer

#TRANSFORM WORD VECTOR TO TF IDF
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
#SAVE TF-IDF
pickle.dump(tfidf_transformer, open("tfidf.pkl","wb"))
```

#### The 1st Model we use is Multinomial Naive Bayes, Splite the data in to train and test dataset, we have a classifier (MultinomialNB() ) to fit the training data and finally save th modele to 'nb_model.pkl' 

```python
# Multinomial Naive Bayes
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_train_tfidf, training_data.flag, test_size=0.22, random_state=15)
clf = MultinomialNB().fit(X_train, y_train)

#SAVE MODEL
pickle.dump(clf, open("nb_model.pkl", "wb"))
```

### we can Test the model 

```python
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

category_list = ["ስፖርት", "ጤና", "ቢዝነስ", "ቴክኖሎጂ", "ትምህርት", "ፖለቲካ", 'utf-8']

new_docs = "የኢኖቬሽንና ቴክኖሎጂ ሚኒስቴር እንዲሁም የውሀ፣ መስኖና ኢነርጂ"
new_docs = [new_docs]

#LOAD MODEL
loaded_vec = CountVectorizer(vocabulary=pickle.load(open("count_vector.pkl", "rb")))
loaded_tfidf = pickle.load(open("tfidf.pkl","rb"))
loaded_model = pickle.load(open("nb_model.pkl","rb"))

X_new_counts = loaded_vec.transform(new_docs)
X_new_tfidf = loaded_tfidf.transform(X_new_counts)
predicted = loaded_model.predict(X_new_tfidf)

print(category_list[predicted[0]])
```

### Run the model on the test dataste, it will show us the "predict and actual data"

```python
predicted = loaded_model.predict(X_test)
result_bayes = pandas.DataFrame( {'true_labels': y_test,'predicted_labels': predicted})
result_bayes.to_csv('pred_bayes.csv', sep = ',')

for predicted_item, result in zip(predicted, y_test):
    print(category_list[predicted_item], ' - ', category_list[result])
```
